```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1, 2, 3, 4, 5]
y=[10, 20, 30, 40, 50]
plt.figure(figsize=(3, 6))
plt.plot(x, y, color='red', linestyle='--', linewidth=3)
plt.title("Пример")
plt.grid()
plt.xlabel("Название оси Х")
plt.ylabel("Название оси Y")
```

```python
x=[1, 2, 3, 4, 5]
y1=[10, 20, 30, 40, 50]
y2=[50, 40, 30, 20, 10]
plt.plot(x, y1, color='red', linestyle='--', linewidth=5)
plt.plot(x, y2, color='brown', linestyle='-', linewidth=3)
plt.title("Пример")
plt.grid()
plt.xlabel("Название оси Х")
plt.ylabel("Название оси Y")
plt.legend(["Первая", "Вторая"], title="Легенда", frameon=True)
```

```python
x=[1, 2, 3, 4, 5]
y=[10, 20, 30, 40, 50]
plt.scatter(x, y, color='red', marker='s', linewidth=2)
plt.title("Пример")
plt.grid()
plt.xlabel("Название оси Х")
plt.ylabel("Название оси Y")
```

```python
labels=['Категория 1', 'Категория 2', 'Категория 3']
sizes=[30, 45, 25]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
```

```python
categories=['Категория 1', 'Категория 2', 'Категория 3']
values=[10, 15, 8]
plt.bar(categories, values)
plt.xlabel("Категории")
plt.ylabel("Значения")
plt.title("Пример столбчатой диаграммы")
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=100, density=True, color='blue', alpha=0.5)
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.title('Пример гистограммы')
```
