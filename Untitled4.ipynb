```python
import matplotlib.pyplot as plt
```

```python
x=[1, 2, 3, 4, 5]
y=[10, 20, 30, 40, 50]
plt.figure(figsize=(3, 6))
plt.plot(x, y, color='red', linestyle='--', linewidth=3)
plt.title("Пример")
plt.grid()
plt.xlabel("Название оси Х")
plt.ylabel("Название оси Y")
```

```python
x=[1, 2, 3, 4, 5]
y1=[10, 20, 30, 40, 50]
y2=[50, 40, 30, 20, 10]
plt.plot(x, y1, color='red', linestyle='--', linewidth=5)
plt.plot(x, y2, color='brown', linestyle='-', linewidth=3)
plt.title("Пример")
plt.grid()
plt.xlabel("Название оси Х")
plt.ylabel("Название оси Y")
plt.legend(["Первая", "Вторая"], title="Легенда", frameon=True)
```

```python
x=[1, 2, 3, 4, 5]
y=[10, 20, 30, 40, 50]
plt.scatter(x, y, color='red', marker='s', linewidth=2)
plt.title("Пример")
plt.grid()
plt.xlabel("Название оси Х")
plt.ylabel("Название оси Y")
```
