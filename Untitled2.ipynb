В данном датафрейме казалось, что очевидная корреляция, какую можно был бы увидеть - это завимисость дохода семьи с общими тратами. Т.е. чем больше средний заработок семьи, тем больше у них трат в сумме по всем группам расходов (транспорт, здоровье, еда, жилье и т.п.), но нет. Зависимость = 0.337 - довольно низкая корреляция.
Коррелируют траты на жилье с другими тратами первой необходимости, = 0.915,  чем больше семья тратится на жилье, тем больше иных трат первой неоходимости.
Также чем больше общих трат у семьи, тем больше семья тратится на еду (корреляция = 0.894)
